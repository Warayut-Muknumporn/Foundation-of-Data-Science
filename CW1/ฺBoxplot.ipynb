```python
# Import the libraries pandas, and matplotlib
# for matplotlib just load pyplot
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt #This is to format your plots and do things like plt.figure(figsize=(20,10))
import numpy as np
#import plotly.express as px
import seaborn as sns
```

```python
filename = "fish1.txt"
#If the data file is not in the same fold of this jupyter code file, 
#the filename should include the path
time_weight_bait_data = pd.read_csv(filename, sep = " ", header = None, index_col = None)
#Give human-readable names to the columns
time_weight_bait_data.columns=["Times(hr.)","Weight(kg.)","Bait"]
```

```python
from matplotlib.patches import Polygon

fig, ax1 = plt.subplots(figsize=(10, 6))
fig.canvas.manager.set_window_title('A Boxplot Example')
fig.subplots_adjust(left=0.075, right=0.95, top=0.9, bottom=0.25)

#bp = ax1.boxplot(time_weight_bait_data, notch=0, sym='+', vert=1, whis=1.5)

#print(time_weight_bait_data["Times(hr.)"])
num_boxes1 = len(time_weight_bait_data)

data = []
data10 = []
datatime = []
dataweight = []

for i in range(num_boxes1):
    #print(i)
    datatime.append(np.floor(time_weight_bait_data["Times(hr.)"][i]))
    dataweight.append(time_weight_bait_data["Weight(kg.)"][i])
    temp = []
    temp.append(np.floor(time_weight_bait_data["Times(hr.)"][i]))
    temp.append(time_weight_bait_data["Weight(kg.)"][i])
    data.append(temp)
    if i < 10:
        data10.append(temp)

data0 = []
data1 = []
data2 = []
data3 = []
data4 = []
data5 = []
data6 = []
data7 = []
data8 = []
data9 = []
data10 = []
data11 = []
data12 = []
data13 = []
data14 = []
data15 = []
data16 = []
data17 = []
data18 = []
data19 = []
data20 = []
data21 = []
data22 = []
data23 = []


for k in range(num_boxes1):
    #print(i)
    tempdata = int(np.floor(time_weight_bait_data["Times(hr.)"][k]))
    if tempdata == 0:
        data0.append(time_weight_bait_data["Weight(kg.)"][k])
    if tempdata == 1:
        data1.append(time_weight_bait_data["Weight(kg.)"][k])
    if tempdata == 2:
        data2.append(time_weight_bait_data["Weight(kg.)"][k])
    if tempdata == 3:
        data3.append(time_weight_bait_data["Weight(kg.)"][k])
    if tempdata == 4:
        data4.append(time_weight_bait_data["Weight(kg.)"][k])
    if tempdata == 5:
        data5.append(time_weight_bait_data["Weight(kg.)"][k])
    if tempdata == 6:
        data6.append(time_weight_bait_data["Weight(kg.)"][k])
    if tempdata == 7:
        data7.append(time_weight_bait_data["Weight(kg.)"][k])
    if tempdata == 8:
        data8.append(time_weight_bait_data["Weight(kg.)"][k])
    if tempdata == 9:
        data9.append(time_weight_bait_data["Weight(kg.)"][k])
    if tempdata == 10:
        data10.append(time_weight_bait_data["Weight(kg.)"][k])
    if tempdata == 11:
        data11.append(time_weight_bait_data["Weight(kg.)"][k])
    if tempdata == 12:
        data12.append(time_weight_bait_data["Weight(kg.)"][k])
    if tempdata == 13:
        data13.append(time_weight_bait_data["Weight(kg.)"][k])
    if tempdata == 14:
        data14.append(time_weight_bait_data["Weight(kg.)"][k])
    if tempdata == 15:
        data15.append(time_weight_bait_data["Weight(kg.)"][k])
    if tempdata == 16:
        data16.append(time_weight_bait_data["Weight(kg.)"][k])
    if tempdata == 17:
        data17.append(time_weight_bait_data["Weight(kg.)"][k])
    if tempdata == 18:
        data18.append(time_weight_bait_data["Weight(kg.)"][k])
    if tempdata == 19:
        data19.append(time_weight_bait_data["Weight(kg.)"][k])
    if tempdata == 20:
        data20.append(time_weight_bait_data["Weight(kg.)"][k])
    if tempdata == 21:
        data21.append(time_weight_bait_data["Weight(kg.)"][k])
    if tempdata == 22:
        data22.append(time_weight_bait_data["Weight(kg.)"][k])
    if tempdata == 23:
        data23.append(time_weight_bait_data["Weight(kg.)"][k])
        
datadis = []
datadis.append(data0)
datadis.append(data1)
datadis.append(data2)
datadis.append(data3)
datadis.append(data4)
datadis.append(data5)
datadis.append(data6)
datadis.append(data7)
datadis.append(data8)
datadis.append(data9)
datadis.append(data10)
datadis.append(data11)
datadis.append(data12)
datadis.append(data13)
datadis.append(data14)
datadis.append(data15)
datadis.append(data16)
datadis.append(data17)
datadis.append(data18)
datadis.append(data19)
datadis.append(data20)
datadis.append(data21)
datadis.append(data22)
datadis.append(data23)

#datadis.append(data0)

#print(data0)
print(datadis)



bp = ax1.boxplot(data, notch=0, sym='+', vert=1, whis=1.5)

plt.setp(bp['boxes'], color='palevioletred')
plt.setp(bp['whiskers'], color='palevioletred')
plt.setp(bp['fliers'], color='red', marker='+')

ax1.yaxis.grid(True, linestyle='-', which='major', color='lightgrey',
               alpha=0.5)

ax1.set(
    axisbelow=True,  # Hide the grid behind plot objects
    title='Comparison of Distributions',
    xlabel='Distribution',
    ylabel='Value',
)
#-----------------------

box_colors = ['darkkhaki', 'royalblue']
num_boxes = len(data)
medians = np.empty(num_boxes)
for i in range(num_boxes):
    box = bp['boxes'][i]
    box_x = []
    box_y = []
    for j in range(5):
        box_x.append(box.get_xdata()[j])
        box_y.append(box.get_ydata()[j])
    box_coords = np.column_stack([box_x, box_y])
    # Alternate between Dark Khaki and Royal Blue
    ax1.add_patch(Polygon(box_coords, facecolor=box_colors[i % 2]))
    # Now draw the median lines back over what we just filled in
    med = bp['medians'][i]
    median_x = []
    median_y = []
    for j in range(2):
        median_x.append(med.get_xdata()[j])
        median_y.append(med.get_ydata()[j])
        ax1.plot(median_x, median_y, 'k')
    medians[i] = median_y[0]
    # Finally, overplot the sample averages, with horizontal alignment
    # in the center of each box
    ax1.plot(np.average(med.get_xdata()), np.average(data[i]),
             color='w', marker='*', markeredgecolor='k')
#---------------------
# Set the axes ranges and axes labels
ax1.set_xlim(0.5, num_boxes + 0.5)
top = 25
bottom = 0
ax1.set_ylim(bottom, top)
#ax1.set_xticklabels(np.repeat(random_dists, 2),
#                    rotation=45, fontsize=8)

# Due to the Y-axis scale being different across samples, it can be
# hard to compare differences in medians across the samples. Add upper
# X-axis tick labels with the sample medians to aid in comparison
# (just use two decimal places of precision)
pos = np.arange(num_boxes) + 1
upper_labels = [str(round(s, 2)) for s in medians]
weights = ['bold', 'semibold']
for tick, label in zip(range(num_boxes), ax1.get_xticklabels()):
    k = tick % 2
    ax1.text(pos[tick], .95, upper_labels[tick],
             transform=ax1.get_xaxis_transform(),
             horizontalalignment='center', size='x-small',
             weight=weights[k], color=box_colors[k])



plt.show()
```
